###Algoritmos detallados de mi programa

1. Algoritmo de "Alta de preguntas de prueba PISA" 
    
    1. Se pide el id de la pregunta.
    2. Se pide el enfoque de la pregunta.
    3. Se pide el texto de la pregunta.
    4. Se piden las opciones de respuesta.
    5. Se pide la respuesta correcta.
    6. se gurda en una lista global.

2. Algoritmo de "Actualizar preguntas de la prueba PISA" 

  1. Se pide el ID de la pregunta que se quiere actualizar 
  2. Se revisa si el Id existe en la lista.
  3. Si existe se le da la opción de cambiar si se escribe actualizar o no cambiar si se presiona enter.
  4. Se vuelven a llenar todos los datos solicitados para guardar una pregunta.
  5. Se pregunta si se quiere actualizar y si se quiere actualizar se actualiza en la lista preguntas

3. Algoritmo de "Estudiar preguntas de cierta área"

  1. Se pide el área que se quiere estudiar.
  2. Se pide la cantidad de preguntas qué se quieren estudiar.
  3. Se busca la cantidad de preguntas en el enfoque elegido.
  4. Se presentan las preguntas con opciones de respuesta.
  5. Se espera que se responda.
  6. Se evalúa si la respuesta es correcta.
  7. Se muestra correcta si es correcta e incorrecta si es incorrecta la respuesta. 
  8. Se repite el proceso hasta que termine el ciclo.

4. Algoritmo de "Presentar un quiz"

  1. Se preparan las preguntas para el quiz.
  2. Se despliega la pregunta con la operacione de respuesta.
  3. Se contesta la pregunta y se guarda en una lista.
  4. Se completa el ciclo al tener 10 preguntas contestadas

5. Algoritmo de "Reporte de calificaciones"

  1. Se muestra el mensaje de que se presentarán los datos solicitados 
  2. Se presentan el total de usuarios que entraron.
  3. Se presenta el promedio del quiz
  4. Se presenta el porcentaje de las preguntas correctas 
  5. Se presenta el porcentaje de las preguntas incorrectas

6. Algoritmo de "Salir"

    1. Se recibe la orden de llamar a la función salir.
    2. Se imprime "hasta pronto".
    3. Se define start como false para que se corte la cadena while.
    4. Se devuelve este valor.

```python
import time #Importé la librería de tiempo

preguntas = [] 

registro = [] #Lista para la función de presentar quiz. Registra los intentos, promedio, buenas y malas del usuario

preguntas_archivo = "preguntas_archivo.txt"


def leer_archivo():

  global preguntas

  with open(preguntas_archivo, "r") as archivo:
    leer = archivo.readlines()

    for elemento in leer:
      variable = elemento.split(",")
      variable[7] = variable[7][:-1]
      preguntas.append(variable)

def grabar_archivo():
  with open(preguntas_archivo, "w") as archivo:
    for pregunta in preguntas:
      texto = ",".join(pregunta)
      archivo.write(texto + "\n")

def menu():

  print(
    """Menú de opciones:
    1.	Alta de preguntas de prueba PISA (lectura, matemáticas, ciencias)
    2.	Actualizar preguntas de la prueba PISA (lectura, matemáticas, ciencias)
    3.	Estudiar preguntas de cierta área (lectura, matemáticas, ciencias)
    4.	Presentar un quiz (lectura, matemáticas, ciencias)
    5.	Reporte de calificaciones
    6.	Salir"""
  )
  opciones = (int(input("Selecciona una accion del menu: ")))
  return opciones 

def salir():
  print("Hasta pronto")
  start = False
  return start #Regresa el valor False a la función Main para terminar el ciclo While

def registrar_pregunta():

  global preguntas
  preguntas=[]
  leer_archivo()

  # Establece los valores para la variable pregunta.
  print("Bienvenido a la opción para registrar preguntas, por favor sigue las instrucciones")
  id_pregunta = str(input("\t\tIngresa el ID de la pregunta: "))
  enfoque = input("\tIngresa el enfoque de la pregunta: Lectura, Matemáticas, Ciencias: ")
  pregunta = input("\t\tIngresa el texto de la pregunta: ")
  respuestas = ""

  # Establece los valores para la variable respuestas.
  for num in range(4):
    respuesta = input("\tIngresa la respuesta " + str(num + 1) + ": ")

    if num == 3:
      respuestas = respuestas + respuesta 
    else:
      respuestas = respuestas + respuesta + "," 
 
  respuestas = respuestas.split(",") 
  
  respuesta1, respuesta2, respuesta3, respuesta4 = respuestas 

  correcta = input("\tIngresa la respuesta correcta:")

  pregunta = [id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta]
  
  # Se guarda la variable pregunta como lista en la lista global preguntas
  preguntas.append(pregunta)
  # Se guarda en el archivo

  grabar_archivo()



def actualizar_pregunta():
 
  global preguntas
  preguntas=[]
  leer_archivo()
  
  print("Bienvenido a la opción para actualizar preguntas, por favor sigue las instrucciones \nA continuación se despliega la lista de preguntas guardadas\n")

  for pregunta in preguntas:
    
    id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = pregunta 

    print(f"ID de la pregunta: {id_pregunta} \nEnfoque de la pregunta: {enfoque} \nPregunta: {pregunta}")
    print(f"a){respuesta1:<15}b){respuesta2:<15}c){respuesta3:<15}d){respuesta4:<15} \n")
    
  id_cambio = input("\tIngresa el id de la pregunta que quieres cambiar: ")

  # Este ciclo busca en cada uno de los renglones de la lista preguntas el id_cambio ingresado.
  posicion = -1
  found_value = False
  for cambio in preguntas:
    posicion += 1
    # Este ciclo if rompe cuando se encuentra el id_cambio
    if id_cambio in cambio:
      found_value=True
      break
  
  #Evita un error para que no se gusrden id de preguntas que no existen 
  if posicion > len(preguntas) or found_value == False:
    print("\tNo se encotro")
  else:
    enter = input("\tSi deseas no deseas cambiar la pregunta presiona enter, y si deseas cambiarla escribe Actualizar: ")

    if enter == "":
      print("\tPregunta no actualizada")

    #Vuelve a registrar la pregunta conmo en la funcion de registrar pregunta
    elif enter != "":
      id_pregunta = input("\t\tIngresa el ID de la pregunta: ")
      enfoque = input("\t\tIngresa el enfoque de la pregunta: Lectura, Matemáticas, Ciencias: ")
      pregunta = input("\t\tIngresa el texto de la pregunta: ")
      respuestas = ""

      for num in range(4):
        respuesta = input("Ingresa la respuesta " + str(num + 1) + ": ")

        if num == 3:
          respuestas = respuestas + respuesta 
        else:
          respuestas = respuestas + respuesta + "," 
    
      respuestas = respuestas.split(",") 
      
      respuesta1, respuesta2, respuesta3, respuesta4 = respuestas 

      correcta = input("Ingresa la respuesta correcta:")

      tpregunta = [id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta]

      print("----------------- Valores de la pregunta modificada -------------------")
      print(f"Enfoque de la pregunta: {enfoque} \nPregunta: {pregunta}")
      print(f"a){respuesta1:<15}b){respuesta2:<15}c){respuesta3:<15}d){respuesta4:<15} \n")
      print("-----------------------------------------------------------------------")

      #Pregunta si realmente se quiere actualizar la pregunta
      actualizar = input(str("\t¿Está seguro de que quiere actualizar los datos de la pregunta? SI/NO: "))

      if actualizar == "SI":
        preguntas[posicion] = tpregunta
        grabar_archivo()
        print("\t\t\tPregunta actualizada.")
      else:
        print("\t\t\tPregunta no actualizada.")


def estudiar_area():

  global preguntas
  preguntas=[]
  leer_archivo()

  enfoque_estudio = input("\tIngresa el enfoque que quieres estudiar: ")

  cantidad_estudio = int(input("\tIngresa  la cantidad de preguntas que se quieren estudiar: "))

  estudio = [] # Nueva lista con las preguntas seleccionadas
  contador = 0 # Indice que lleva la cuenta de las preguntas
  
  # Llenda la lista de estudio
  for pregunta in preguntas:
    # Valida el enfoque para llenar la lista de estudio
    if enfoque_estudio in pregunta:
      contador += 1  # Incrementa el indice de preguntas 
      estudio.append(pregunta) # Inserta la pregunta en la lista de estudio
      # Valida que el contaror no sobre pase la cantidad de preguntas de estudio
      if (contador >= cantidad_estudio):
        break
 
 
  # Ciclo que se encargara de mostrar las preguntas de la 
  for estudiar in estudio:

    id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = estudiar
    print(f"Pregunta: {pregunta}")
    print(f"a){respuesta1:<15}b){respuesta2:<15}c){respuesta3:<15}d){respuesta4:<15} \n")
    respuesta_estudio = input("Tu respuesta es: ")

    # Valida si la pregunta es correcta o no e imprime si es correcta o incorrecta
    if respuesta_estudio == correcta:
      print("\t\tcorrecta")
    else:
      print("\t\tincorrecta")

def presentar_quiz():

  global registro
  global preguntas
  preguntas=[]
  leer_archivo()

  cantidad_estudio = 10  #Determina la cantidad de preguntas totales
  estudio = []  #Lista donde se gardan las preguntas
  quiz = []  #Lista donde se registran datos del quiz
  contador = 0  

  #Evita que se pase de 10 preguntas
  for pregunta in preguntas:
    contador = contador + 1
      #print("Pregunta %s %s" % (enfoque_estudio, pregunta))
    estudio.append(pregunta)

    if (contador >= cantidad_estudio):
      break

  # Determina el tiempo
  tinicio = time.perf_counter()
  tfin = time.perf_counter() ## Este es el cronometro
  print("\t\tTiempo de inicio %s" % tinicio)

  for estudiar in estudio:

    id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = estudiar

    print("\t\tTiempo transcurrido minutos %d" % ((tfin-tinicio)/60))
    print(f"Enfoque {enfoque} \nPregunta: {pregunta}")
    print(f"a){respuesta1:<15}b){respuesta2:<15}c){respuesta3:<15}d){respuesta4:<15} \n")

    respuesta_quiz = input("\t\t\tTu respuesta es: ")
    cuenta_correcta=0
    registra_quiz = []


    if respuesta_quiz == correcta:
      cuenta_correcta = 1
    registra_quiz = [id_pregunta, enfoque, pregunta, cuenta_correcta]

    tfin = time.perf_counter()

    if ((tfin - tinicio)/60>20):
      print("\tSe termino tu tiempo, la última pregunta no será tomada en cuenta para la calificación")
      break    

    quiz.append(registra_quiz)
 
  print("\t\tInició %s  y termino %s" % (tinicio, tfin))
  
  intento = len(registro)

  malas = 0
  buenas = 0

  for q in quiz:
    no_pregunta, enfoque, pregunta, correcta = q

    if correcta == 1:
      buenas = buenas + 1
    if correcta == 0:
      malas = malas + 1
      
  prom = buenas / 10
  cardex = [intento+1, prom, malas, buenas] #Se llena la variable global llamada registro
  registro.append(cardex)

  print("\tRegistro de intentos: %s" % registro)

def reporte_calificaciones():
  global registro
  intentos = 0
  tbuenas = 0
  tmalas = 0
  tprom = 0

  #Permite determinar los valores para los intentos, promedio buenas y malas
  for r in registro:
    intento, prom, malas, buenas = r
    intentos = intentos + 1
    tprom = tprom + prom
    tbuenas = tbuenas + buenas
    tmalas = tmalas + malas

  total_usuarios = intentos
  promedio = tprom / intentos
  preguntas_correctas = (tbuenas*100) / (intentos*10)
  preguntas_incorrectas = (tmalas*100) / (intentos*10)

  print(f"Intentos {intentos} \nPromedio: {promedio} \nPromedio de preguntas buenas: {preguntas_correctas} \nPromedio de preguntas malas: {preguntas_correctas}")


def main():

  start = True

  while start:

    opciones = menu()

    if opciones == 1:
      registrar_pregunta()

    elif opciones == 2:
      actualizar_pregunta()

    elif opciones == 3:
      estudiar_area()

    elif opciones == 4:
      presentar_quiz()

    elif opciones == 5:
      reporte_calificaciones()

    elif opciones == 6:
      start = salir()

    else:
      print("Opción incorrecta")

main()
```
